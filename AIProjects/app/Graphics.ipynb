```python
import matplotlib.pyplot as plt
import random
```

### 单画布多坐标系

```python
fig, axes = plt.subplots(figsize=(20, 8),nrows=1, ncols=2, sharex=True)
x = range(60)
y = range(30)
labels = ['上海市十一到十二点的温度', '北京市十一到十二点的温度']
x_tics = ['11点{}分'.format(i) for i in x]
y_shanghai = [random.uniform(13, 23) for i in x]
y_beijing = [random.uniform(3, 8) for i in x]
axes[0].plot(x, y_shanghai, 'r+', linestyle='-', label=labels[0])
axes[1].plot(x,y_beijing, 'b*', linestyle='--', label=labels[1])
plt.xticks(x[::5], x_tics[::5])
plt.yticks(y[::5])
for i in range(2):
    axes[i].set_yticks(y[::5])
    axes[i].set_xlabel('时间')
    axes[i].set_ylabel('温度')
    axes[i].set_title(labels[i])
    axes[i].legend(loc='best')

plt.show()
```

### 饼状图

```python
plt.figure(figsize=(10, 8), dpi=200)
movie_name = ['雷神3: 诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴','降魔传','追捕','七十七天','密战','狂兽','其它']
place_tic = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6764,52222]
plt.pie(place_tic, labels=movie_name, autopct='%.2f%%', colors=['b','r','g','y','c','m','y','g','c','r','g'])
plt.axis('equal')
plt.show()
```

### 爆裂
