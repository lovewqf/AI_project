```python
import numpy as np
```

```python
a1 = np.array(['a', 'b', 'c'])
a2 = np.array([['L', 'H'], ['L', 'L']])
a3 = np.array([ [[1,2,3],[3,4,5]],   [[5,6,6],[7,8,9]] ])
```

### 基本属性shpae

```python
print('a1:', a1.shape)
print('a2:', a2.shape)
print('a3', a3.shape)
```

### 维度ndim

```python
print('a1:', a1.ndim)
print('a2:', a2.ndim)
print('a3', a3.ndim)
```

### 元素数量size

```python
print('a1:', a1.size)
print('a2:', a2.size)
print('a3', a3.size)
```

### 每一个元素所占的字节itemsize

```python
print('a1:', a1.itemsize)
print('a2:', a2.itemsize)
print('a3', a3.itemsize)
```

### 所占的总字节数nbytes

```python
print('a1:', a1.nbytes)
print('a2:', a2.nbytes)
print('a3', a3.nbytes)
```

### flags

```python
print('a1:', a1.flags)
print('a2:', a2.flags)
print('a3', a3.flags)
```

### 数据类型

```python
a1.dtype, a2.dtype, a3.dtype
```

### 类型转换

```python
a = np.array([[1,2,3], [4,5,6]], dtype=np.float32)
```
