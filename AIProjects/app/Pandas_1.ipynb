### 以某列的值作为新的行索引

```python
df = pd.DataFrame({'month': [1,4,7,10], 'year': [1,1,2,2], 'sale': [55,40,84,31]})
df
```

```python
# 设置列变为行索引
df.set_index(['month'])
```

```python
# 设置多重索引,具有多重索引的结构,相当于三维数组结构
# 通常会使用MultiIndex这种结构解决三维数据表示问题
df = df.set_index(['year', 'month'])
df
```

```python
# 相当于三维数组结构,通常会使用MultiIndex这种结构解决三维数据表示问题
df.index
```

## 基本操作: Series,DataFrame的索引完成切片获取

### 1.索引操作

```python
# 使用某个股票每天的数据
data = pd.read_csv('D:/AI/data/stock_day/stock_day.csv')
data
```

```python
data.head()
```

```python
data['open']
```

```python
# 通过索引访问,先列后行
data['open']['2018-02-26']
```

```python
data[['open', 'high', 'close']]
```

### 结合loc或iloc应用索引

### 通过loc

```python

data.loc['2018-02-27': "2018-02-23", 'open':'close']
```

```python
data.loc['2018-02-27': "2018-02-23", ['open','close']]
```

### 使用iloc

```python
data.iloc[0:3, 0]
```

```python
data.iloc[0:3, :4]
```

### 使用ix(混合)进行下标和名称索引

```python
data.ix[0:4, ['open', 'close', 'low']]
```

## 对内容进行操作

```python
# 修改
data['open']['2018-02-26'] = 26.3
```

```python
data.head()
```

```python
# 修改一列
data['open'] = 1.2
```

```python
data.head()
```

## 排序

### a.对索引进行排序

```python
data.sort_index(ascending=True)
```

### b.对内容进行排序

```python
# 对p_change涨幅进行排序
# ascending: True从小到大,False从大到小,默认为True
data.sort_values(by='p_change', ascending=False).head()
```

```python
# 多字段时使用列表,先按照第一个键排,相同时在按照第二个键排
data.sort_values(by=['high', 'low']).head()
```
