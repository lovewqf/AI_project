```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
plt.figure(1, figsize=(9, 3))
plt.subplot(1,3,1)
plt.bar(names, values)
plt.subplot(1,3,2)
plt.scatter(names, values,c= 'r')
plt.subplot(1,3,3)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
# 
plt.figure(figsize=(10, 6), dpi=200)
x = range(60)
y = [random.uniform(15, 18) for i in x]
x_ch = ['11{}'.format(i) for i in x]
y_ticks = range(40)
plt.plot(x, y, 'r*', linestyle='--', label='lin1')
# 1,2
plt.xticks(x[: : 5], x_ch[: : 5])
plt.yticks(y_ticks[::5])
plt.title('XX')
plt.legend(loc='best')
plt.show()
```
