# pandas

```python
import pandas as pd # 数据处理
import numpy as np # 数据运算
import matplotlib.pyplot as plt  # 可视化绘图
# import random
```

```python
# 参数一正态期望,参数二标准差,数据的shape
stock_day_rise = np.random.normal(0, 1, [500, 504])
```

```python
stock_day_rise
```

## pandas的三大数据结构Series, Dataframe,Panel

### 1.DataFrame

```python
# 将数据转成Dataframe结构
stock_df = pd.DataFrame(stock_day_rise)
stock_df
```

```python
print(type(stock_df))
print(np.shape(stock_df))
```

### 2.DataFrame既有行索引又有列索引

```python
# 获取某支股票
np.shape(stock_df)[0]
```

```python
# 获取某个交易日
np.shape(stock_df)[1]
```

```python
stock_code = ['股票'+str(i) for i in range(np.shape(stock_df)[0])]
stock_code
```

```python
pd.DataFrame(stock_day_rise, index=stock_code)
```

```python
# 给定开始日期创建时间序列
pd.date_range('2017-01-01', periods=504)
```

```python
pd.date_range('2017-01-01', periods=np.shape(stock_day_rise)[1], freq='B')
```
