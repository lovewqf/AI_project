3.构造数据显示

```python
# 画出温度变化图
plt.figure(figsize=(10, 6), dpi=200)
x = range(60)
y = [random.uniform(15, 18) for i in x]
x_ch = ['11点{}分'.format(i) for i in x]
y_ticks = range(40)
plt.plot(x, y, 'r*', linstyle=)
# 参数1设置步长,参数2设置每个点对应的文字描述
plt.xticks(x[: : 5], x_ch[: : 5])
plt.yticks(y_ticks[::5])
plt.show()
```

7.再添加一个城市的温度变化

```python
plt.figure(figsize=(10, 6), dpi=200)
x = range(60)
y_shanghai = [random.uniform(12, 18) for i in x]
y_beijing = [random.uniform(3, 8) for i in x]
x_ch = ['11点{}分'.format(i) for i in x]
y_ticks = range(40)
plt.plot(x, y_shanghai)
plt.plot(x, y_beijing)
# 参数1设置步长,参数2设置每个点对应的文字描述
plt.xticks(x[: : 5], x_ch[: : 5])
plt.yticks(y_ticks[::5])
plt.title('一些城市的温度显示')
plt.xlabel('时间')
plt.ylabel('温度')
plt.show()
```

8.自定义线型

```python
plt.figure(figsize=(10, 6), dpi=200)
x = range(60)
y_shanghai = [random.uniform(12, 18) for i in x]
y_beijing = [random.uniform(3, 8) for i in x]
x_ch = ['11点{}分'.format(i) for i in x]
y_ticks = range(40)
plt.plot(x, y_shanghai, 'r*', linestyle='--')
plt.plot(x, y_beijing, 'b+', linestyle=':')
# 参数1设置步长,参数2设置每个点对应的文字描述
plt.xticks(x[: : 5], x_ch[: : 5])
plt.yticks(y_ticks[::5])
plt.title('一些城市的温度显示')
plt.xlabel('时间')
plt.ylabel('温度')
plt.show()
```

9.添加注释

```python
plt.figure(figsize=(10, 6), dpi=200)
x = range(60)
y_shanghai = [random.uniform(12, 18) for i in x]
y_beijing = [random.uniform(3, 8) for i in x]
x_ch = ['11点{}分'.format(i) for i in x]
y_ticks = range(40)
plt.plot(x, y_shanghai, 'r*', linestyle='--', label='上海')
plt.plot(x, y_beijing, 'b+', linestyle=':', label='北京')
# 参数1设置步长,参数2设置每个点对应的文字描述
plt.xticks(x[: : 5], x_ch[: : 5])
plt.yticks(y_ticks[::5])
plt.title('一些城市的温度显示')
plt.xlabel('时间')
plt.ylabel('温度')
plt.legend(loc='best')
plt.show()
```
