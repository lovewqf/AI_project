```python
import matplotlib.pyplot as plt
```

### 多项单列对比

```python
plt.figure(figsize=(20, 8), dpi=300)
movie_name = ['雷神3: 诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴','降魔传','追捕','七十七天','密战','狂兽','其它']
y = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6764,52222]
x = range(len(movie_name))
plt.bar(x,y, width=0.5, color=['b','r','g','y','c','m','y','k','c','r','g'])
plt.xticks(x, movie_name)
plt.xlabel('影片名')
plt.ylabel('票房')
plt.show()
```

### 单项双列对比

```python
plt.figure(figsize=(20, 8), dpi=200)
movie_name = ['雷神3: 诸神黄昏','正义联盟','寻梦环游记']
first_day = [44521, 10321, 20321]
first_week = [32110, 21250, 45155]
x = range(len(movie_name))
plt.bar(x,first_day,width=0.2, label='first')
plt.bar([i+0.2 for i in x],first_week, width=0.2, label='first_week')
plt.xticks([i+0.1 for i in x], movie_name)
plt.xlabel('影片名')
plt.ylabel('票房')
plt.legend(loc='best')
plt.show()
```
