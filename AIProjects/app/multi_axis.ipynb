```python
import matplotlib.pyplot as plt
import random
```

10.同一画布,不同坐标系显示
注意数据写入和坐标系的设置有所不同

```python
x = range(60)
y_ticks = range(40)
# 创建多坐标系
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), dpi=200, sharex=True)
y_shanghai = [random.uniform(12, 18) for i in x]
y_beijing = [random.uniform(3, 8) for i in x]
x_ch = ['11点{}分'.format(i) for i in x]
plt.xlim(0, 60)
axes[0].plot(x, y_shanghai, 'r*', linestyle='--', label='上海')
axes[1].plot(x, y_beijing, 'b+', linestyle=':', label='北京')
# 参数1设置步长,参数2设置每个点对应的文字描述
plt.xticks(x[: : 5], x_ch[: : 5])
axes[0].set_yticks(y_ticks[::5])
axes[1].set_yticks(y_ticks[::5])
axes[0].set_title('中午11点到12点之间的温度变化')
axes[1].set_title('中午11点到12点之间的温度变化')
axes[0].set_xlabel('时间')
axes[0].set_ylabel('温度')
axes[1].set_xlabel('时间')
axes[1].set_ylabel('温度')
axes[0].legend(loc='best')
axes[1].legend(loc='best')
plt.show()
```
