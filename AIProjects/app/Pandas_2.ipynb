### 综合分析: describe()

```python
# 列的详细信息统计分析
data.describe()
```

```python
# 单独获取一个信息字段,0代表列,1代表行
# count(),mean(),std(),min(),max()
data.max(0)
```

```python
data = data.sort_index()
```

```python
data.head()
```

```python
# cumsum()序列求和,axis=0时对列求和(默认),axis=1时对行求和
data.cumsum()
```

```python
# 对某一列求和
data.p_change.cumsum()
```

```python
# plot方法集成了前面直方图,条形图,饼图,折线图
plt.figure(figsize=(20,8),dpi=100)
data.p_change.cumsum().plot()
plt.show()
```
