1.数据读取存储API

1.1read_csv和to_csv

文件的读取和存储

pandas.read_csv(filepath_or_buffer,sep=',',delimiter=None)

filepath_or_buffer:文件路径

usecols:指定读取的列名,列表形式

```python
# 文件的读取和存储
# 指定usecols来读取部分数据,其它不用到的就不需要读取了
# stock_day = pd.read_csv("./data/stock_day/stock_day.csv")
stock_day = pd.read_csv("./data/stock_day/stock_day.csv",usecols=['open','close','high'])
```

```python
stock_day.head()
# stock_day
```

保存

DataFrame.to_csv(path_or_buf=None,sep=',',columns=None,header=True,index=True,index_label=None,mode="w",encoding=None)

path_or_buf:string or file handle,default None

sep:character,dafault','

columns:sequence,optional

mode:'w':重写,'a'追加

index:是否写进行索引

header:boolean or list of string, default True,是否写进列索引值

```python
# index:存储不会讲索引值变成一列数据
# mode:'a'追加方式
# header:已追加方式时候,记住要把列名去除header=False
# stock_day[:10].to_csv("./test.csv",columns=['open'])
```
