4.显示刻度

```python
# 画出温度变化图
# 1.fig
plt.figure(figsize=(20, 8), dpi=80)

# 准备x, y坐标的数据,x,y 一一对应
x = range(60)
y = [random.uniform(15, 18) for i in x]
y_ticks = range(40)

# 显示温度变化

#折线图绘制
plt.plot(x,y)

# 显示刻度， ：5间隔，步长
plt.xticks(x[::5])
plt.yticks(y_ticks[::5])

plt.show()
```

7.再添加一个城市的温度变化

```python
# 画出温度变化图
# 1.fig
plt.figure(figsize=(20, 8), dpi=80)

# 准备x, y坐标的数据,x,y 一一对应
x = range(60)

# 上海
y_shanghai = [random.uniform(15, 18) for i in x]

#北京
y_beijing = [random.uniform(1, 3) for i in x]

x_ch = ["11点{}分".format(i) for i in x]
y_ticks = range(40)

# 显示温度变化

#上海折线图绘制
plt.plot(x,y_shanghai)

#北京拆线图绘制
plt.plot(x,y_beijing)

# 显示刻度， ：5间隔，步长
plt.xticks(x[::5],x_ch[::5])

plt.yticks(y_ticks[::5])

plt.xlabel("时间")
plt.ylabel("温度")
plt.title("一些城市的温度变化曲线图")

plt.show()
```

8.自定义类型

```python
# 画出温度变化图
# 1.fig
plt.figure(figsize=(20, 8), dpi=80)

# 准备x, y坐标的数据,x,y 一一对应
x = range(60)

# 上海
y_shanghai = [random.uniform(15, 18) for i in x]

#北京
y_beijing = [random.uniform(1, 3) for i in x]

x_ch = ["11点{}分".format(i) for i in x]
y_ticks = range(40)

# 显示温度变化

#上海折线图绘制
plt.plot(x,y_shanghai, label="上海")

#北京拆线图绘制
plt.plot(x,y_beijing,color='r', linestyle='--', label="北京")

# 显示刻度， ：5间隔，步长
plt.xticks(x[::5],x_ch[::5])

plt.yticks(y_ticks[::5])

plt.xlabel("时间")
plt.ylabel("温度")
plt.title("一些城市的温度变化曲线图")

plt.show()
```

9.添加注释

```python
# 画出温度变化图
# 1.fig
plt.figure(figsize=(20, 8), dpi=80)

# 准备x, y坐标的数据,x,y 一一对应
x = range(60)

# 上海
y_shanghai = [random.uniform(15, 18) for i in x]

#北京
y_beijing = [random.uniform(1, 3) for i in x]

x_ch = ["11点{}分".format(i) for i in x]
y_ticks = range(40)

# 显示温度变化

#上海折线图绘制
plt.plot(x,y_shanghai, label="上海")

#北京拆线图绘制
plt.plot(x,y_beijing,color='r', linestyle='--', label="北京")

# 显示刻度， ：5间隔，步长
# 指定显示的x刻度的列表, 
# 第一个参数：必须是值
# 第二个参数：指定跟第一个参数对应的中文
# y 0 ~ 40
# plt:对于整体处理，ax:对于每个坐标系处理
plt.xticks(x[::5],x_ch[::5])

plt.yticks(y_ticks[::5])

plt.xlabel("时间")
plt.ylabel("温度")
plt.title("一些城市的温度变化曲线图")

plt.legend(loc="best")
plt.show()
```
